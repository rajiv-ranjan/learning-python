## **Print( )**

the **print(**objects**)** function is used to print(display) a python object(s) in Python as standard output..

It can be used to print strings, numbers, variables, or even the result of expressions.


Syntax: print(*objects, sep=' ', end='\n', file=None, flush=False)

```python
print("Hello everyone!")
```

You can also print multiple items by separating them with commas:

```python
#Printing multiple objects
name = "Adithiya"
age = 21
print("My name is", name, "and I am", age, "years old.")
```

For more control over formatting, you can use f-strings or the format() method for string formatting:

```python
# Using f-strings
print(f"My name is {name} and I am {age} years old.")

# Using format() method
print("My name is {} and I am {} years old.".format(name, age))
```

### Aditional arguments for **print( )**

**sep** (separator): This argument specifies the string to separate multiple arguments. By default, it's a space.

```python
print("Hello", "Participants", sep="-")
```

**end**: This argument defines the string to print at the end, replacing the default newline character.

```python
print("I am", end=" ") # space is giving as end
print("Varun")
```

```python
#When we print two objects separately on consecutive lines, the output will be printed on two lines.
#This happens because the end argument is set as "\n" by default, where "\n" represents a newline character, causing the output to be printed on the next line.
print("I am")
print("Varun")
```

 ## **open( )**

 the **open( )** function is used to open a file and return a corresponding file object.

 This function provides a flexible way to work with files of various types and modes (e.g., read mode, write mode, binary mode, etc.).

Syntax: open(file, mode='r', buffering=-1, encoding=None, errors=None, newline=None, closefd=True, opener=None)


* **file**: The path to the file that needs to be opened. It can be a relative or absolute file path.

* **mode**: Specifies the mode in which the file is opened. It can be '**r**' for reading (default), '**w**' for writing, '**a**' for appending, '**b**' for binary mode, '**t**' for text mode (default), etc.

* For additional information about the open() function and its arguments(buffering, encoding, errors, newline, closefd, opener), please go through its documentation thoroughly.

* [open()](https://docs.python.org/3/library/functions.html#print)

```python
# Open a file for reading
with open('File_data.txt', 'r') as file:
    # Read the contents of the file
    content = file.read()
    # Print the content
    print(content)
```

```python
# Open the file for reading
with open('File_data.txt', 'r') as file:
    # Iterate over the file line by line
    for line in file:
        # Split each line into name and grade
        name, grade = line.strip().split(',')
        # Convert grade to integer
        grade = int(grade)
        # Check if the grade is passing
        if grade >= 60:
            # Print the name of the first student who passed
            print("passing student:", name)
            # Exit the loop after finding the first passing student
            pass
    else:
        # If no passing student is found, print a message
        print("end")
```

## **input( )**

The input() function is used to get user input during the execution of your Python program.

It pauses the program's execution and waits for the user to enter some data that we desire to input.




```python
name = input("What is your name? ")  #If the prompt argument is present, it is written to standard output without a trailing newline.
print(f"Hello, {name}!")
```

Example:

Let's write a function that calculates the BMI of the coder by getting their weight and height from them.

```python
def bmi_calculator():
    # Input weight in kilograms
    weight = float(input("Enter your weight in kilograms: "))

    # Input height in meters
    height = float(input("Enter your height in meters: "))

    # Calculate BMI
    bmi = weight / (height ** 2)

    # Print BMI
    print("Your BMI is:", bmi)

# Call the function to execute the BMI calculation
bmi_calculator()
```
