# **list()**
**list(**iterable**)** Converts an iterable (such as a tuple or string) into a list

```python
#This example converts a tuple (1, 2, 3, 4) into a list
tuple_data = (1, 2, 3, 4)
list_data = list(tuple_data)
print(list_data)
```

When you apply the **list()** function to a string, it converts each character of the string into individual elements of a list

```python
list("hello world")
```

# **tuple()**
**tuple(**iterable**)** Converts an iterable (such as a list or string) into a tuple

```python
# This example converts a list [1, 2, 3, 4] into a tuple
list_data = [1, 2, 3, 4]
tuple_data = tuple(list_data)
print(tuple_data)
```

Similar to list(), tuple() converts each character of a string into individual elements of a tuple

```python
tuple("string")
```

 # **set()**
 **set(**iterable**)** Creates a set from an iterable (such as a list, tuple, or string).

```python
#This example converts a list [1, 2, 2, 3, 4, 4, 5] into a set, removing duplicate elements.
list_data = [1, 2, 2, 3, 4, 4, 5]
set_data = set(list_data)
print(set_data)
```

 # **dict()**
 **dict()** is a built-in constructor function in Python used to create dictionary objects.

 It can be used in various ways, such as providing **keyword arguments**, **passing a mapping object**, or **providing an iterable of key-value pairs**.

## Method 1:  keyword arguments
syntax: dict(**kwarg)

If keyword arguments are used, the arguments must be valid identifiers as they become keys in the dictionary.

```python
# Using keyword arguments
my_dict1 = dict(one=1, two=2, three=3)
print(my_dict1)
```

## Method 2:  Using a mapping object
syntax: dict(mapping, **kwarg)

If a mapping object is provided, it must be a dictionary or a subclass of a dictionary.

```python
my_dict2 = dict({'one': 1, 'two': 2, 'three': 3})
print(my_dict2)
```

## Method 2:  Using an iterable of key-value pairs
syntax: dict(iterable, **kwarg)

If an iterable of key-value pairs is provided, each element of the iterable must be a tuple where the first element is the key and the second element is the value.

```python
my_dict3 = dict([('one', 1), ('two', 2), ('three', 3)])
print(my_dict3)
```
